```python
air_quality = {}
for idx, quality in zip(range(1,len(df['AQI_Bucket'].unique())+1), df['AQI_Bucket'].unique()):
    air_quality[quality]= idx
air_quality
```

```python
df["Air_quality"]= df["AQI_Bucket"].map(air_quality)
```

```python
df
```

```python
city_idx = {}
for idx, city_n in zip(range(1,len(df['City'].unique())+1), df['City'].unique()):
    city_idx[city_n]= idx
city_idx
```

```python
df['City']=df['City'].map(city_idx)
```

```python
df
```

```python
df = df.drop(["AQI_Bucket", "Date"], axis = 1)
```

```python
df['id'] = list(range(1,6237))
```

```python
df = df.set_index('id')
```

```python
x = df.drop(["AQI","Air_quality"], axis = 1)
y = df["AQI"]
y2 = df["Air_quality"]
```

```python
x.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_scaled = sc.fit_transform(x)
x_scaled[:2]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=35)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc = RandomForestRegressor(n_estimators=100,max_depth = 10,random_state= 10).fit(X_train,y_train)
```

```python
y_pred =rfc.predict(X_test)
```

```python
y_pred
```

```python
rfc.score(X_test,y_test)
```
